### Преобразуем тип данных столбцов **`scounted_price, actual_price, discount_percentage, rating, rating_count`** из **`object`** в числовой.

```python
df['discounted_price'] = df['discounted_price'].str.replace("₹",'')
df['discounted_price'] = df['discounted_price'].str.replace(",",'')
df['discounted_price'] = df['discounted_price'].astype('float64')

df['actual_price'] = df['actual_price'].str.replace("₹",'')
df['actual_price'] = df['actual_price'].str.replace(",",'')
df['actual_price'] = df['actual_price'].astype('float64')

df['discount_percentage'] = df['discount_percentage'].str.replace('\D','', regex=True).astype('float64') / 100 # Делим на 100, так как в этом столбцы проценты

df['rating_count'] = df['rating_count'].str.replace(",",'')
df['rating_count'] = df['rating_count'].astype('float64')
```

```python
# У столбца rating есть значение, не являющиеся числовым значением 
df['rating'].unique()
```

```python
# Заменим | на nan
df.loc[df['rating'] == '|','rating'] = np.nan
```

```python
df['rating'] = df[df['rating'].notna()]['rating'].str.replace('\D','', regex=True).astype('float64')
df['rating'] = df['rating'].apply(lambda x: x / 10 if x > 5 else x) 
```

```python
df['rating_count'].fillna(round(df['rating_count'].mean()), inplace=True)
df['rating'].fillna(round(df['rating'].mean()), inplace=True)
df.loc[[282, 324, 1279],:]
```

### Преобразование данных и их визуализация

### Столбец **`category`** содержит категории товара, упорядаченные по основную категорию и подкатегории. Оставим только основную и первую подкатегорию

```python
categories = df['category'].str.split('|', expand=True)
categories
```

```python
df['main_category'] = categories.copy().iloc[:,0]
df['sub_category'] = categories.copy().iloc[:,1]
df.drop('category', inplace=True, axis=1)
df
```

```python
maincategory_count = pd.DataFrame(df.groupby(['main_category'])['discounted_price'].count())
maincategory_count.rename(columns={'discounted_price':'count'}, inplace=True)
maincategory_count
```

```python
maincategory_count.reset_index().sort_values(by='count').head(5)
```

```python
sns.barplot(data=maincategory_count.reset_index().sort_values(by='count',ascending=False).head(5), x='count', y='main_category', palette='pastel')
plt.title('Количество товаров по категориям', fontweight='heavy')
plt.show()
```

### Видим, что самые распространенные категории товаров - это  **`Electronics`**,**`Computers&Accessories`** и **`Home&Kitchen`**, у которых около 500 товаров.
### Потом идет **`OfficeProducts`** с 31 товаром. У остальных категорий в данном наборе данных меньше 3 товаров.

```python
subcategory_count = pd.DataFrame(df.groupby(['main_category', 'sub_category'])['discounted_price'].count())
subcategory_count.rename(columns={'discounted_price':'count'}, inplace=True)
subcategory_count
```

```python
sns.barplot(data=subcategory_count.reset_index().sort_values(by='count',ascending=False).head(5), x='count', y='sub_category', palette='pastel')
plt.title('Количество товаров по подкатегориям', fontweight='heavy')
plt.show()
```

```python
pop_cats = ['Computers&Accessories','Electronics','Home&Kitchen','OfficeProducts']

fig, ax = plt.subplots(4, figsize=(12, 12)) 
plt.subplots_adjust(hspace = 0.7)
fig.suptitle('Количество товаров в подкатегориях для каждой основной категории', fontweight='heavy', size='x-large')

for axs in range(4):
    mask = subcategory_count.reset_index()['main_category'].isin([pop_cats[axs]])
    sns.barplot(data=subcategory_count.reset_index()[mask], y='count', x='sub_category', ax=ax[axs], palette='pastel')
    ax[axs].set_xlabel('')
    ax[axs].set_title(pop_cats[axs],  fontweight='heavy')
    for tick in ax[axs].get_xticklabels():
        tick.set_rotation(20)
         
        

plt.show()
```

Это млэет свидетельствать о том, что большинство пользователей не желают указывать своего имени при написании отзыва


## Цены и Скидочные цены

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

sns.histplot(x=df.actual_price,ax=ax[0], kde=True, bins=15, color='blue')
sns.histplot(x=df.discounted_price,ax=ax[1], kde=True, bins=15, color='red')

ax[0].set_title('Распределение реальных цен',  fontweight='heavy')
ax[0].set_xlabel('Реальная цена',  fontweight='bold')
ax[0].set_ylabel('Кол-во',  fontweight='bold')

ax[1].set_title('Распределение цен cо скидкой',  fontweight='heavy')
ax[1].set_xlabel('Цена со скидкой',  fontweight='bold')
ax[1].set_ylabel('Кол-во', fontweight='bold')

ax[0].axvline(df['actual_price'].mean(), label=f'Среднее {int(df.actual_price.mean())}')
ax[0].legend()

ax[1].axvline(df['discounted_price'].mean(), label=f'Среднее {int(df.discounted_price.mean())}')
ax[1].legend()

plt.show()
```

Распредения реальных цен и цен со скидкой схожи. Они скошены в левую часть и крутовершина.

```python
temp = [df['actual_price'].kurt(), df['actual_price'].skew(), df['discounted_price'].kurt(), df['discounted_price'].skew()]
print(f'Эксцесс: реальные цены - {temp[0]}, цены со скидкой {temp[2]}')
print(f'Ассиметрия: реальные цены - {temp[1]}, цены со скидкой {temp[3]}')
```

## Рейтинг и количество оценок

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15,5))

sns.histplot(x=df.rating,ax=ax[0], kde=True, bins=15, color='blue')
sns.histplot(x=df.rating_count,ax=ax[1], kde=True, bins=15, color='red')

ax[0].set_title('Распределение оценки товаров',  fontweight='heavy')
ax[0].set_xlabel('Оценка',  fontweight='bold')
ax[0].set_ylabel('Кол-во',  fontweight='bold')

ax[1].set_title('Распределение количества оценок на один товар',  fontweight='heavy')
ax[1].set_xlabel('Количество оценок на один товаров',  fontweight='bold')
ax[1].set_ylabel('Число товаров с определеном количество товаров', fontweight='bold')

ax[0].axvline(df['rating'].mean(), label=f'Медиана {round(df.rating.median(),2)}')
ax[0].legend()

ax[1].axvline(df['rating_count'].mean(), label=f'Среднее {int(df.rating_count.mean())}')
ax[1].legend()

plt.show()
```

Видим, что в подавляющем числе случаев оценка выше трех.
